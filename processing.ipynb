```python
kode = df['Kode Identitas'].astype(str).str.ljust(16, 'X')

df['kab']     = kode.str[:4]
df['kec']     = kode.str[:7]
df['desa']    = kode.str[:10]
df['sls']     = kode.str[:14]
df['sub_sls'] = kode.str[:16]

df[['Kode Identitas', 'kab', 'kec', 'desa', 'sls', 'sub_sls']].head()
```

```python
levels = ['kab', 'kec', 'desa', 'sls', 'sub_sls']

tabulasi = {}
for lvl in levels:
    tab = (
        df.groupby([lvl, 'Skala Usaha / Jenis Prelist'])
        .size()
        .unstack(fill_value=0)
    )
    tab['Total'] = tab.sum(axis=1)
    tabulasi[lvl] = tab

# Tampilkan semua tabulasi
for lvl, tab in tabulasi.items():
    print(f"\n{'='*60}")
    print(f"Tabulasi berdasarkan: {lvl.upper()}")
    print('='*60)
    display(tab)
```
